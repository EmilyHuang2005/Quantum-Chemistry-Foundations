## Integrals

A Gaussian function

\begin{equation}
\chi_\mu(\mathbf{r}) = N_{\alpha_\mu} e^{-\alpha_\mu r^2}
\end{equation}

Some fundamental integrals of a Gaussian ($x \in [0,\infty)$)

\begin{align}
I_0(\alpha) &= \int_0^\infty dx \; e^{-\alpha x^2} = \frac{1}{2} \sqrt{\frac{\pi}{\alpha}}
\\
I_1(\alpha) &= \int_0^\infty dx \; x e^{-\alpha x^2} = \frac{1}{2 \alpha}
\\
I_2(\alpha) &= \int_0^\infty dx \; x^2 e^{-\alpha x^2} = \frac{1}{4\alpha} \sqrt{\frac{\pi}{\alpha}}
\\
I_4(\alpha) &= \int_0^\infty dx \; x^4 e^{-\alpha x^2} = \frac{3}{8\alpha^2} \sqrt{\frac{\pi}{\alpha}}
\end{align}

### Normalization

Let's compute the integral $\langle \chi_\mu | \chi_\nu \rangle $, defined as
\begin{align}
\langle \chi_\mu | \chi_\nu \rangle  & = \int d\mathbf{r} \; \chi^*_\mu(\mathbf{r}) \chi_\nu(\mathbf{r}) = N^*_{\alpha_\mu} N_{\alpha_\nu}
\int d\mathbf{r} \; e^{-(\alpha_\mu + \alpha_\nu) r^2} \\
& = N^*_{\alpha_\mu} N_{\alpha_\nu}
4\pi \underbrace{\int_0^\infty dr \; r^2  e^{-(\alpha_\mu + \alpha_\nu) r^2}}_{I_2(\alpha_\mu + \alpha_\nu)} \\
& = N^*_{\alpha_\mu} N_{\alpha_\nu}
\left(\frac{\pi}{\alpha_\mu + \alpha_\nu}\right)^{3/2}
\end{align}

From the normalization condition ($\langle \chi_\mu | \chi_\mu \rangle = 1$) it follows that
\begin{equation}
N_{\alpha_\mu} = 
\left(\frac{2\alpha_\mu}{\pi}\right)^{3/4}
\end{equation}

From this result we get
\begin{align}
\langle \chi_\mu | \chi_\nu \rangle  & = \int d\mathbf{r} \; \chi^*_\mu(\mathbf{r}) \chi_\nu(\mathbf{r}) = N^*_{\alpha_\mu} N_{\alpha_\nu}
\int d\mathbf{r} \; e^{-(\alpha_\mu + \alpha_\nu) r^2} \\
& = N^*_{\alpha_\mu} N_{\alpha_\nu}
4\pi \int_0^\infty dr \; r^2  e^{-(\alpha_\mu + \alpha_\nu) r^2} \\
& = \frac{2^{3/2} \alpha_\mu^{3/4} \alpha_\nu^{3/4} }{(\alpha_\mu + \alpha_\nu)^{3/2}}
\end{align}

### Kinetic energy integrals

We can proceed similarly to evaluate the potential energy integrals

\begin{align}
\langle \chi_\mu | \hat{T} | \chi_\nu \rangle  & = -\frac{1}{2} \int d\mathbf{r} \; \chi^*_\mu(\mathbf{r}) \nabla^2 \chi_\nu(\mathbf{r}) =  -\frac{1}{2} N^*_{\alpha_\mu} N_{\alpha_\nu}
\int d\mathbf{r} \;  e^{-\alpha_\mu r^2} \frac{1}{r^2} \frac{\partial}{\partial r} \left(r^2 \frac{\partial}{\partial r} e^{-\alpha_\nu r^2} \right)
\end{align}

\begin{equation}
\frac{1}{r^2} \frac{\partial}{\partial r} \left(r^2 \frac{\partial}{\partial r} e^{-\alpha_\nu r^2} \right)
= \frac{1}{r^2} \frac{\partial}{\partial r} \left(-2 \alpha_\nu r^3 e^{-\alpha_\nu r^2} \right)
= (-6 \alpha_\nu  + 4 \alpha_\nu^2 r^2) e^{-\alpha_\nu r^2} 
\end{equation}

\begin{align}
\langle \chi_\mu | \hat{T} | \chi_\nu \rangle  & =  -\frac{1}{2} N^*_{\alpha_\mu} N_{\alpha_\nu}
\int d\mathbf{r} \;  e^{-\alpha_\mu r^2} (-6 \alpha_\nu  + 4 \alpha_\nu^2 r^2) e^{-\alpha_\nu r^2}  \\
& =  -\frac{1}{2} \left[
-6 \alpha_\nu \langle \chi_\mu | \chi_\nu \rangle
+ 4 \alpha_\nu^2 \langle \chi_\mu | r^2 | \chi_\nu \rangle
\right] \\
& = 
3 \alpha_\nu \langle \chi_\mu | \chi_\nu \rangle
- 2 \alpha_\nu^2 \langle \chi_\mu | r^2 | \chi_\nu \rangle
\end{align}


Let's compute the integral $\langle \chi_\mu |  r^2 | \chi_\nu \rangle $
\begin{align}
\langle \chi_\mu | r^2 |\chi_\nu \rangle  & = \int d\mathbf{r} \; \chi^*_\mu(\mathbf{r})r^2 \chi_\nu(\mathbf{r}) = N^*_{\alpha_\mu} N_{\alpha_\nu}
\int d\mathbf{r} \; r^2 e^{-(\alpha_\mu + \alpha_\nu) r^2} \\
& = N^*_{\alpha_\mu} N_{\alpha_\nu}
 \int_0^\infty dr \; r^4  e^{-(\alpha_\mu + \alpha_\nu) r^2} \\
& = N^*_{\alpha_\mu} N_{\alpha_\nu} 
\frac{3 \pi^{3/2}}{2 (\alpha_\mu + \alpha_\nu)^{5/2}} \\
& = \sqrt{2} \frac{3 \alpha_\mu^{3/4} \alpha_\nu^{3/4}}{(\alpha_\mu + \alpha_\nu)^{5/2}} \\
\end{align}

Finally we get (after some algebra)
\begin{align}
\langle \chi_\mu | \hat{T} | \chi_\nu \rangle 
& = 3 \alpha_\nu \langle \chi_\mu | \chi_\nu \rangle
- 2 \alpha_\nu^2 \langle \chi_\mu | r^2 | \chi_\nu \rangle \\
& = 6 \sqrt{2}  \frac{ \alpha_\mu^{7/4} \alpha_\nu^{7/4}}{(\alpha_\mu + \alpha_\nu)^{5/2}}
\end{align}

### Potential energy integrals

We can proceed similarly to evaluate the potential energy integrals

\begin{align}
\langle \chi_\mu | \hat{V} | \chi_\nu \rangle  & = -\int d\mathbf{r} \; \chi^*_\mu(\mathbf{r}) \frac{1}{r} \chi_\nu(\mathbf{r}) = -N^*_{\alpha_\mu} N_{\alpha_\nu}
\int d\mathbf{r} \; \frac{1}{r} e^{-(\alpha_\mu + \alpha_\nu) r^2} \\
& = -N^*_{\alpha_\mu} N_{\alpha_\nu}
4\pi \int_0^\infty dr \; r  e^{-(\alpha_\mu + \alpha_\nu) r^2} \\
& = -N^*_{\alpha_\mu} N_{\alpha_\nu}  
\left(\frac{2\pi}{\alpha_\mu + \alpha_\nu}\right) \\
& = -4 \sqrt{\frac{2}{\pi}}  \frac{\alpha_\mu^{3/4} \alpha_\nu^{3/4} }{\alpha_\mu + \alpha_\nu}
\end{align}
